```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
x,y=make_classification(n_samples=1000,n_features=2,n_informative=2,n_redundant=0,n_clusters_per_class=1)
```

```python
for i in range(2):
  row_ix=np.where(y==i)
  a,b=x[row_ix,0],x[row_ix,1]
  plt.scatter(a,b)
plt.show()
```

```python
from sklearn.cluster import DBSCAN,AffinityPropagation,KMeans,OPTICS,AgglomerativeClustering
```

##Affinity Propagation - Intituitioin

```python
AP=AffinityPropagation(damping=0.9)
AP.fit(x)
y_hat=AP.predict(x)
```

```python
clusters=np.unique(y_hat)
```

```python
for cluster in clusters:
  row_ix=np.where(y_hat==cluster)
  plt.scatter(x[row_ix,0],x[row_ix,1])
plt.show()
```

```python
def plot_clusters(clusters,x,y):
  for cluster in clusters:
    row_ix=np.where(y==cluster)
    plt.scatter(x[row_ix,0],x[row_ix,1])
  plt.show()
```

##Agglomerative clustering - Intituition

```python
Agglo_clustering=AgglomerativeClustering(n_clusters=2)
y_hat=Agglo_clustering.fit_predict(x)
clusters=np.unique(y_hat)
plot_clusters(clusters,x,y_hat)
```

##BIRCH Intituition - Balanced Iteration Reducing and Clustering using Hierarchies

```python
from sklearn.cluster import Birch
```

```python
birch=Birch(threshold=0.1,n_clusters=2)
```

```python
birch.fit(x)
y_hat=birch.predict(x)
clusters=np.unique(y_hat)
plot_clusters(clusters,x,y_hat)
```

##DBSCAN Intituition - Density based spatial clustering of Applications with Noise

```python
dbscan=DBSCAN(eps=0.15,min_samples=9)
```

```python
y_hat=dbscan.fit_predict(x)
clusters=np.unique(y_hat)
plot_clusters(clusters,x,y_hat)
```

## KMeans - Clustering

```python
kmeans_c=KMeans(n_clusters=2)
```

```python
kmeans_c.fit(x)
y_hat=kmeans_c.predict(x)
clusters=np.unique(y_hat)
plot_clusters(clusters,x,y_hat)
```

##MiniBatch Kmeans clustering

```python
from sklearn.cluster import MiniBatchKMeans
```

```python
MBKmeans = MiniBatchKMeans(n_clusters=2)
```

```python
MBKmeans.fit(x)
y_hat=MBKmeans.predict(x)
clusters=np.unique(y_hat)
plot_clusters(clusters,x,y_hat)
```

## Meanshift - ###
Mean shift clustering involves finding and adapting centroids based on the density of examples in the feature space.

```python
from sklearn.cluster import MeanShift
```

```python
meanshift=MeanShift()
```

```python
y_hat=meanshift.fit_predict(x)
clusters=np.unique(y_hat)
plot_clusters(clusters,x,y_hat)
```

## OPTICS - Ordering points to indentify clustering structure##
It does not produce a clustering of a data set explicitly; but instead creates an augmented ordering of the database representing its density-based clustering structure. This cluster-ordering contains information which is equivalent to the density-based clusterings corresponding to a broad range of parameter settings.



```python
optics=OPTICS(max_eps=2,min_samples=9)
```

```python
y_hat=optics.fit_predict(x)
clusters=np.unique(y_hat)
print(clusters)
plot_clusters(clusters,x,y_hat)
```

##Spectual Clustering 
A promising alternative that has recently emerged in a number of fields is to use spectral methods for clustering. Here, one uses the top eigenvectors of a matrix derived from the distance between points.

```python
from sklearn.cluster import SpectralClustering
```

```python
spectral_c=SpectralClustering(n_clusters=2)
```

```python
y_hat=spectral_c.fit_predict(x)
clusters=np.unique(y_hat)
print(len(clusters))
plot_clusters(clusters,x,y)
```

## Gaussian Mixture Clustering
A Gaussian mixture model summarizes a multivariate probability density function with a mixture of Gaussian probability distributions as its name suggests.

```python
from sklearn.mixture import GaussianMixture
```

```python
gaussian_clustering=GaussianMixture(n_components=2)
```

```python
y_hat=gaussian_clustering.fit_predict(x)
clusters=np.unique(y_hat)
plot_clusters(clusters,x,y_hat)
```
